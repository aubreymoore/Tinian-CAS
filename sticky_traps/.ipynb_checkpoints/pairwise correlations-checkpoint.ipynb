```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mean_coordinates2.csv')
df
```

```python
# recode scale_level values from ['L', 'M', 'H'] to [1, 2, 3] to enable correlation analysis
df['scale_level'].replace(['L', 'M', 'H'], [1, 2, 3], inplace=True)
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(
    df[['scale_level', 'crawlers','coccinellids']].corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
# heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
# save heatmap as .png file
# dpi - sets the resolution of the saved image in dots/inches
# bbox_inches - when set to 'tight' - does not allow the labels to be cropped
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```
